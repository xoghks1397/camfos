```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("aaa.csv",encoding = 'euc-kr')
```

```python
df['홈페이지'] = df['홈페이지'].fillna('없음')
df['전화'] = df['전화'].fillna('없음')
df['예약 구분'] = df['예약 구분'].fillna('없음')
```

```python
for i in df.index.values:
    if df.iloc[i]['홈페이지'][0] != 'h' and df.iloc[i]['홈페이지'] != '없음' :
        df.loc[(df.컨텐츠ID == i+1),'홈페이지'] = 'http://' + df.loc[(df.컨텐츠ID == i+1),'홈페이지']
```

```python
for i in df.index.values:
    str = ''
    if df.iloc[i]['홈페이지'] != '없음' :
        str += '온라인, '
    if df.iloc[i]['전화'] != '없음' :
        str += '전화'
    
    else:
        str = '현장'
    
    df.loc[(df.컨텐츠ID == i+1),'예약 구분'] = str
```
